```python
import skimage
import skimage.draw
import skimage.io
import skimage.transform
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
plt.figure(figsize=(20,20))

for n in range(1):
    I = np.zeros((512,512))

    for i in range(10000):
        x = np.random.randint(0,512)
        y = np.random.randint(0,512)
        
        r = np.random.randint(4,7)
        cr,cc = skimage.draw.circle(x,y,r,(512,512))
        I[cr,cc] += 0.1
        
#     cr,cc = skimage.draw.circle(100,100,100,(1024,1024))
#     I[cr,cc] = 0.8
#     I = skimage.transform.resize(I,(512,512),order=3,anti_aliasing=True)
#     skimage.io.imsave('C:/Users/charl/Dropbox/simulatedBeadsImages_0324/%d.png' % n, (I*255).astype('uint8') )
    print('Done image',n)
    plt.imshow(I)
plt.show()
```
